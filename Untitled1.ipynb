```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split

#X = df[['CLOSING_QTY', 'CLOSEING_VALUE', 'DIVISION_CODE', 'DUMP', 'OPEN_QTY', 'OPEN_VALUE', 'PARTY_CODE', 'PRODUCT_CODE', 'RECEIPT_QTY', 'RECEIPT_VALUE']]
X = df[[ 'OPEN_QTY', 'PARTY_CODE', 'PRODUCT_CODE', 'RECEIPT_QTY']]
y = df['ISSUE_QTY']

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Predict on the testing data
y_pred = model.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor

decision_tree_regressor = DecisionTreeRegressor(
    criterion='squared_error',
    max_depth=5,
    min_samples_split=2,
    min_samples_leaf=1,
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    max_features=None)

decision_tree_regressor.fit(X_train, y_train)
y_pred_dt = decision_tree_regressor.predict(X_test)

md=mean_squared_error(y_test, y_pred_dt)
print('mean_squared_error is :',md)

mad=mean_absolute_error(y_test, y_pred_dt)
print('mean_absolute_error is :',mad)

rd=r2_score(y_test, y_pred_dt)
print('r2_score is :',rd)
```
